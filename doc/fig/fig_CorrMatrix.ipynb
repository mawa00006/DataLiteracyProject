```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl

from tueplots import bundles
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("../../dat/preprocessed_brca_metabric_clinical_data.tsv", sep='\t')

data = data[['Age at Diagnosis', 'Neoplasm Histologic Grade', 'Lymph nodes examined positive', 'Mutation Count', 'Tumor Size', 'Tumor Stage', 'Nottingham prognostic index', 'Overall Survival (Months)', 'Relapse Free Status (Months)']]

ticks = ['Age at Diagnosis', 'Histologic Grade', 'Lymph nodes positive', 'Mutation Count', 'Tumor Size', 'Tumor Stage', 'NPI', 'Overall Survival', 'Relapse Free Status']
```

```python
plt.rcParams.update(bundles.icml2022(column="half", nrows=1.35, ncols=1))

cols = data.columns
corr = data[cols].corr()

fig, ax = plt.subplots()
sns.heatmap(corr, xticklabels=ticks, yticklabels=ticks, cmap=mpl.colors.LinearSegmentedColormap.from_list("", ['red','white','steelblue']), center=0.0)

plt.savefig('fig_CorrMatrix.pdf')
```
